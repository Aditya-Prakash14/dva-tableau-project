# 02 Cleaning


```python
# Loading of the Dataset
import pandas as pd

df = pd.read_csv('../data/raw/vehicles_raw.csv')

print("Shape:", df.shape)
print(df.columns)
```

```python
# Top 5 rows
df.head()
```

```python
# Last 5 rows
df.tail()
```

```python
# Null values and its counts
df.isnull().sum()
```

```python
# Duplicate removals
df = df.drop_duplicates(subset=['id'], keep='first')
```

```python
# Cleaning cylinders column and formatting
# Extracting numeric values
df['cylinders'] = df['cylinders'].str.extract(r'(\d+)')
df['cylinders'] = pd.to_numeric(df['cylinders'], errors='coerce')
```

```python
# Converting to numeric columns
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['odometer'] = pd.to_numeric(df['odometer'], errors='coerce')
```

```python
# Handling numeric missing columns

# Drop rows where target variable is missing
df = df.dropna(subset=['price'])

# Fill important numeric features
df['year'] = df['year'].fillna(df['year'].median())
df['odometer'] = df['odometer'].fillna(df['odometer'].median())
df['cylinders'] = df['cylinders'].fillna(df['cylinders'].mode()[0])

# Drop unused geo columns
df = df.drop(columns=['lat', 'long'])
```

```python
# Dropping irrelavant columns
df = df.drop(columns=[
    'url', 'region_url', 'image_url',
    'county', 'VIN'
], errors='ignore')
```

```python
# Handling categorical missing values

# Droping weak / redundant columns
df = df.drop(columns=['region', 'size'], errors='ignore')


# Paint color → keep but avoid bias
df['paint_color'] = df['paint_color'].fillna('unknown')

# TEMP fill model for grouping only
df['model'] = df['model'].fillna('unknown_temp')

# -------------------------------
# Smart Imputation (BEFORE unknown fill)
# -------------------------------

# Fuel
fuel_map = df.groupby('model')['fuel'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
df['fuel'] = df['fuel'].fillna(df['model'].map(fuel_map))
df['fuel'] = df['fuel'].fillna(df['fuel'].mode()[0])

# Transmission
trans_map = df.groupby('model')['transmission'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
df['transmission'] = df['transmission'].fillna(df['model'].map(trans_map))
df['transmission'] = df['transmission'].fillna(df['transmission'].mode()[0])

# Type
type_map = df.groupby('model')['type'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
df['type'] = df['type'].fillna(df['model'].map(type_map))
df['type'] = df['type'].fillna(df['type'].mode()[0])

# Drive (model → type → fallback)
drive_map_model = df.groupby('model')['drive'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
df['drive'] = df['drive'].fillna(df['model'].map(drive_map_model))

drive_map_type = df.groupby('type')['drive'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
df['drive'] = df['drive'].fillna(df['type'].map(drive_map_type))

df['drive'] = df['drive'].fillna(df['drive'].mode()[0])


# Identity columns
df['manufacturer'] = df['manufacturer'].fillna('unknown')
df['model'] = df['model'].fillna('unknown')


# Independent columns
df['condition'] = df['condition'].fillna(df['condition'].mode()[0])
df['title_status'] = df['title_status'].fillna(df['title_status'].mode()[0])
df['state'] = df['state'].fillna(df['state'].mode()[0])
df['model'] = df['model'].replace('unknown_temp', 'unknown')
```

```python
# Since we arent using text NLP dropping text column
df = df.drop(columns=['description'], errors='ignore')
```

```python
# Outlier handling

# Price filtering
df = df[df['price'] >= 1500]
upper_price = df['price'].quantile(0.99)
df = df[df['price'] <= upper_price]


# Year filtering
max_year = df['year'].max()
df = df[(df['year'] >= 1980) & (df['year'] <= max_year)]


# Odometer filtering (refined)
lower_odo = max(500, df['odometer'].quantile(0.01))
upper_odo = df['odometer'].quantile(0.99)

df = df[(df['odometer'] >= lower_odo) & (df['odometer'] <= upper_odo)]


# Cylinders validation
df = df[df['cylinders'].isin([3, 4, 5, 6, 8, 10, 12])]
```

```python
# Convert to datetime (force conversion)
df['posting_date'] = pd.to_datetime(df['posting_date'], errors='coerce', utc=True)

# NOW extract features
df['post_year'] = df['posting_date'].dt.year
df['post_month'] = df['posting_date'].dt.month

# Optional: drop rows where date is missing
df = df.dropna(subset=['post_year', 'post_month'])

# Drop original column
df = df.drop(columns=['posting_date'])
```

```python
print("Final Shape:", df.shape)

print("\nNull Values:\n", df.isnull().sum())

print("\nSummary:\n", df.describe())
```

# Feature Engineering

```python
# Using age instead of raw year for car age
current_year = df['post_year'].max()
df['car_age'] = current_year - df['year']
```

```python
# Raw price is hard to interpret visually added cateogy column
def price_bucket(x):
    if x < 5000:
        return 'Budget'
    elif x < 15000:
        return 'Mid-range'
    else:
        return 'Premium'

df['price_category'] = df['price'].apply(price_bucket)
```

```python
# Mileage Category
def mileage_bucket(x):
    if x < 50000:
        return 'Low'
    elif x < 150000:
        return 'Medium'
    else:
        return 'High'

df['mileage_category'] = df['odometer'].apply(mileage_bucket)
```

```python
# Age category
def age_bucket(x):
    if x <= 3:
        return 'New'
    elif x <= 10:
        return 'Moderate'
    else:
        return 'Old'

df['age_category'] = df['car_age'].apply(age_bucket)
```

```python
# Manufacturing Groups
# Too many brands → messy charts
top_brands = df['manufacturer'].value_counts().nlargest(10).index

df['manufacturer_grouped'] = df['manufacturer'].apply(
    lambda x: x if x in top_brands else 'other'
)
```

```python
# Can identify Posting Season using months extracted previously
def get_season(month):
    if month in [12,1,2]:
        return 'Winter'
    elif month in [3,4,5]:
        return 'Spring'
    elif month in [6,7,8]:
        return 'Summer'
    else:
        return 'Fall'

df['posting_season'] = df['post_month'].apply(get_season)
```
