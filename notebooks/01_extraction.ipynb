# 01 Extraction

This notebook is used to load the cleaned "used cars dataset" and perform the initial extraction and profiling steps.

## Objectives
- Load the dataset
- Inspect shape, columns, and data types
- Measure missing values and duplicates
- Check for obvious outliers
- Explore price distribution and key feature relationships

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_theme(style='whitegrid')
```

```python
df = pd.read_csv('../data/raw/vehicles_raw_sampled.csv', low_memory=False)

# Display first few rows to preview dataset
df.head()
```

```python
# Check number of rows, columns, and column names
df.shape, df.columns.tolist()
```

```python
# Inspect data types and non-null counts
df.info()
```

```python
# Calculate missing values count and percentage per column
missing = df.isnull().sum().sort_values(ascending=False)
missing_pct = (df.isnull().mean() * 100).sort_values(ascending=False)

# Shows columns with missing values only
pd.DataFrame({
    'missing_count': missing,
    'missing_percent': missing_pct.round(2)
}).query('missing_count > 0')
```

```python
# Count total duplicate rows
df.duplicated().sum()
```

```python
# Get statistical summary of price column
if 'price' in df.columns:
    df['price'].describe()
```

```python
# Select available numeric columns
numeric_cols = ['price', 'year', 'odometer']
available_numeric = [col for col in numeric_cols if col in df.columns]

# Get statistical summary of numeric features
df[available_numeric].describe()
```

```python
# Count unique values in each column
df.nunique().sort_values(ascending=False)
```

```python
# Identify categorical columns
categorical_cols = df.select_dtypes(include='object').columns

# Get summary of categorical features
df[categorical_cols].describe()
```

```python
# Check duplicates based on key numeric features
df.duplicated(subset=['price', 'year', 'odometer']).sum()
```

```python
# Detect outliers in numeric columns using IQR method
for col in available_numeric:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(f"{col}: {len(outliers)} outliers")
```

```python
# Count number of columns per data type
df.dtypes.value_counts()
```

```python
# Check skewness of price distribution
df['price'].skew()
```
